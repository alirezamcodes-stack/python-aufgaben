# Entwickeln Sie ein Struktogramm fuer folgende Aufgabenstellung: Nach Eingabe von drei Zahlen soll der hoechste Wert dieser Zahlen ausgegeben werden.

```python
max_number = 0  # Anfang

a = int(input("Bitte Num 1 eingeben: "))  # Erste Zahl
b = int(input("Bitte Num 2 eingeben: "))  # Zweite Zahl
c = int(input("Bitte Num 3 eingeben: "))  # Dritte Zahl

max_number = a  # Aktuelles Maximum

if b > max_number:
    max_number = b  # Neues Maximum

if c > max_number:
    max_number = c  # Neues Maximum

print("Hoechste Nummer ist:", max_number)  # Ausgabe
```

# Es soll ein Algorithmus entwickelt werden, welcher den Speicherbedarf fuer digitale Bilder berechnet und in MiB ausgibt. Eingabewerte sind: Anzahl der Bilder, Hoehe und Breite eines einzelnen Bildes in Pixeln und die Farbtiefe pro Pixel in Bit. Erstellen Sie fuer den Algorithmus ein Struktogramm.

```python
anzahl_bilder = int(input("Bitte Anzahl Bilder eingeben: "))  # Bildanzahl
hoehe = int(input("Bitte Hoehe eingeben: "))  # Hoehe
breite = int(input("Bitte Breite eingeben: "))  # Breite
farbtiefe = int(input("Bitte Farbtiefe eingeben: "))  # Farbtiefe

pixel = hoehe * breite  # Pixelzahl
bitmenge = pixel * farbtiefe  # Bitmenge
byte = bitmenge / 8  # Byte
gesamt = byte * anzahl_bilder  # Gesamtmenge
mib = gesamt / (1024 * 1024)  # MiB
print("MiB ist:", mib)  # Ausgabe
```

```python
def checkPassword(password):
    specialchar = False
    letter = False
    digit = False

    if len(password) < 10:
        print("Your password length is short\n")
        return False

    for ch in password:
        if ch.isdigit():
            digit = True
        elif ch.isalpha():
            letter = True
        else:
            specialchar = True

    if letter and digit and specialchar:
        print(f"Your password length ={len(password)}\ndigit={digit}\nspecialchar={specialchar}\n")
        return password
    else:
        print(f"Your password length ={len(password)}\ndigit={digit}\nspecialchar={specialchar}\n")
        print("Try again")
        return False


checkPassword("Alirez!")
print("-------------------------")
checkPassword("Alirez123%$$")
print("-------------------------")
checkPassword("Alir2623623623")
print("-------------------------")
```

```python
from random import randint

def password_generator(x):
    password = []

    for i in range(x):
        j = randint(1, 4)

        if j == 1:
            rnd_select = randint(1,4)
            if rnd_select == 1:
                rnd_specChar = randint(33, 47)
            elif rnd_select == 2:
                rnd_specChar = randint(58, 64)
            elif rnd_select == 3:
                rnd_specChar = randint(91, 96)
            else:
                rnd_specChar = randint(123, 126)
            password.append(chr(rnd_specChar))

        elif j == 2:
            rnd_number = randint(48, 57)
            password.append(chr(rnd_number))

        elif j == 3:
            rnd_bigChar = randint(65, 90)
            password.append(chr(rnd_bigChar))

        else:
            rnd_smallChar = randint(97, 122)
            password.append(chr(rnd_smallChar))

    return "".join(password)

print("""
 в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•— в–Ҳв–Ҳв•—     в–Ҳв–Ҳв•—в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•— в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•— в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—
в–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•—в–Ҳв–Ҳв•‘     в–Ҳв–Ҳв•‘в–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•—в–Ҳв–Ҳв•”в•җв•җв•җв•җв•қв•ҡв•җв•җв–Ҳв–Ҳв–Ҳв•”в•қв–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•—
в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•‘в–Ҳв–Ҳв•‘     в–Ҳв–Ҳв•‘в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•”в•қв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—    в–Ҳв–Ҳв–Ҳв•”в•қ в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•‘
в–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•‘в–Ҳв–Ҳв•‘     в–Ҳв–Ҳв•‘в–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•—в–Ҳв–Ҳв•”в•җв•җв•қ   в–Ҳв–Ҳв–Ҳв•”в•қ  в–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•‘
в–Ҳв–Ҳв•‘  в–Ҳв–Ҳв•‘в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—в–Ҳв–Ҳв•‘в–Ҳв–Ҳв•‘  в–Ҳв–Ҳв•‘в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—в–Ҳв–Ҳв•‘  в–Ҳв–Ҳв•‘
в•ҡв•җв•қ  в•ҡв•җв•қв•ҡв•җв•җв•җв•җв•җв•җв•қв•ҡв•җв•қв•ҡв•җв•қ  в•ҡв•җв•қв•ҡв•җв•җв•җв•җв•җв•җв•қв•ҡв•җв•җв•җв•җв•җв•җв•қв•ҡв•җв•қ  в•ҡв•җв•қ

в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—  в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•— в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—в–Ҳв–Ҳв•—    в–Ҳв–Ҳв•— в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•— в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•— в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—
в–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•—в–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•—в–Ҳв–Ҳв•”в•җв•җв•җв•җв•қв–Ҳв–Ҳв•”в•җв•җв•җв•җв•қв–Ҳв–Ҳв•‘    в–Ҳв–Ҳв•‘в–Ҳв–Ҳв•”в•җв•җв•җв–Ҳв–Ҳв•—в–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•—в–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•—
в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•”в•қв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•‘в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•—в–Ҳв–Ҳв•‘ в–Ҳв•— в–Ҳв–Ҳв•‘в–Ҳв–Ҳв•‘   в–Ҳв–Ҳв•‘в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•”в•қв–Ҳв–Ҳв•‘  в–Ҳв–Ҳв•‘
в–Ҳв–Ҳв•”в•җв•җв•җв•қ в–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•‘в•ҡв•җв•җв•җв•җв–Ҳв–Ҳв•‘в•ҡв•җв•җв•җв•җв–Ҳв–Ҳв•‘в–Ҳв–Ҳв•‘в–Ҳв–Ҳв–Ҳв•—в–Ҳв–Ҳв•‘в–Ҳв–Ҳв•‘   в–Ҳв–Ҳв•‘в–Ҳв–Ҳв•”в•җв•җв–Ҳв–Ҳв•—в–Ҳв–Ҳв•‘  в–Ҳв–Ҳв•‘
в–Ҳв–Ҳв•‘     в–Ҳв–Ҳв•‘  в–Ҳв–Ҳв•‘в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•‘в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•‘в•ҡв–Ҳв–Ҳв–Ҳв•”в–Ҳв–Ҳв–Ҳв•”в•қв•ҡв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•”в•қв–Ҳв–Ҳв•‘  в–Ҳв–Ҳв•‘в–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв–Ҳв•”в•қ
в•ҡв•җв•қ     в•ҡв•җв•қ  в•ҡв•җв•қв•ҡв•җв•җв•җв•җв•җв•җв•қв•ҡв•җв•җв•җв•җв•җв•җв•қ в•ҡв•җв•җв•қв•ҡв•җв•җв•қ  в•ҡв•җв•җв•җв•җв•җв•қ в•ҡв•җв•қ  в•ҡв•җв•қв•ҡв•җв•җв•җв•җв•җв•қ

            рҹ”җ Alireza Password Generator рҹ”җ
""")





numm_pass = int(input("How many passwords do you want to generate? "))
length_pass = int(input( "what length should they be ? "))
for i in range(numm_pass):
    print(f"{i+1}- {password_generator(length_pass)}")

```

```python
def encrypt_text(text):
    encrypt_text = ""

    for ch in text:
        encrypt_text += chr(ord(ch) + 5)

    return encrypt_text


def decrypt_text(text2):
    decrypt_text = ""

    for ch in text2:
        decrypt_text += chr(ord(ch) - 5)

    return decrypt_text


text = input("Pls enter Text: ")
result = encrypt_text(text)
print(result)

text2 = input("Pls enter encrypted Text: ")
result2 = decrypt_text(text2)
print(result2)
```
